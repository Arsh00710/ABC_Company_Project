```python
#step5:Exploratory Data Analysis (EDA) and Visualization


# 1.How many are there in each Team and the percentage splitting with respect to the total employees.

# Count the number of employees in each team

team_counts = data['Team'].value_counts()
team_counts
```

```python
# Calculate the percentage splitting with respect to the total employees
team_percentage = round((team_counts / len(data)) * 100,2)
team_percentage
```

```python
# Create a new DataFrame with the results
team_table = pd.DataFrame({'Team': team_counts.index, 'Number of Employees': team_counts.values, 'Percentage': team_percentage.values})
team_table
```

```python
# The table offers a comprehensive overview of employee distribution across various teams, detailing the employee count within each team and the corresponding percentage share relative to the total workforce. Notably, the "New Orleans Pelicans" boast the highest number of employees, accounting for approximately 4.15% of the total workforce. Conversely, teams such as the "Minnesota Timberwolves" and "Orlando Magic" exhibit lower percentages, each constituting around 3.06% of the total employee count. This data provides valuable insights for organizational planning and resource allocation, facilitating informed decision-making in managing human resources effectively.
```

```python
# Plotting the bar chart with Seaborn

plt.figure(figsize=(12, 6))
sns.barplot(x='Team', y='Percentage', data=team_table, palette='viridis')
plt.title('Percentage of Employees in Each Team')
plt.xlabel('Team')
plt.ylabel('Percentage')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Data Story:
#The visualization presents a breakdown of employee distribution across various teams within ABC Company, offering a snapshot of the workforce composition.

#Team Sizes:
#The data reveals that the New Orleans Pelicans boast the largest workforce with 19 employees, closely followed by the Memphis Grizzlies with 18 employees. This suggests a notable presence of personnel within these teams.

#Percentage Distribution:
#Each team's percentage representation denotes its contribution to the overall employee count. The New Orleans Pelicans stand out with the highest percentage at 4.15%, indicating a substantial share of the company's workforce.

#Consistency:
#Several teams maintain a consistent employee count, hovering around 15 individuals. Examples include the Brooklyn Nets, Portland Trail Blazers, and Oklahoma City Thunder, resulting in comparable percentage values.

#Variation:
#Conversely, certain teams such as the Orlando Magic and Minnesota Timberwolves exhibit a slightly lower percentage of 3.06%, indicating a smaller fraction of the total employee pool.

#Significance of New Orleans Pelicans:
#The New Orleans Pelicans emerge as a significant force within the organization, commanding the highest percentage share. Potential factors driving their sizable presence may include involvement in major projects, strategic initiatives, or serving as a pivotal department within the company.

#Comparison of Orlando Magic & Minnesota Timberwolves:
#Both the Orlando Magic and Minnesota Timberwolves demonstrate similar proportions, each representing 3.06% of the total workforce. This similarity in percentages suggests comparable contributions from these teams, albeit at a lower scale compared to others.
```

```python
# 3.Find from which age group most of the employees belong to.

#To view values_Count  in the 'Age' column of DataFrame for Find from which age group most of the employees belong to.
age_distribution = data['Age'].value_counts().reset_index()
age_distribution.columns = ['Age', 'No of Employees']
age_distribution
```

```python
# Their are 47 employees aged 24, 46 employees aged 25.
```

```python
#Plotting the age distribution using countplot

plt.figure(figsize=(12, 6))
sns.countplot(x='Age', data=data)
plt.title('Age Distribution of Employees')
plt.xlabel('Age')
plt.ylabel('Number of Employees')
plt.show()
```

```python
#Age group analysis
age_groups = pd.cut(data['Age'], bins=[15, 20, 25, 30,35, 40, 45], labels=['15-20', '20-25', '25-30', '30-35', '35-40','40-45'])
age_group_counts = age_groups.value_counts().reset_index()
age_group_counts.columns=['Age group','Number of Employees']
age_group_counts
```

```python
# Plotting the age group distribution using barplot

plt.figure(figsize=(10, 6))
sns.barplot(x='Age group', y='Number of Employees', data=age_group_counts, palette='colorblind')
plt.title('Distribution of Employees Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.show()
```

```python
# 4.Find out under which team and position, spending in terms of salary is high.

# Group the salary by Team and Position, and calculate the sum of salaries for each group
team_position_salary = data.groupby(['Team', 'Position'])['Salary'].sum().reset_index()
team_position_salary 
```

```python

# Team and position, spending in terms of salary is high.
max_total_salary_index = team_position_salary['Salary'].idxmax()
max_total_salary_row = team_position_salary.loc[max_total_salary_index]
max_total_salary_row
```

```python

#Display in descending order of salary
team_position_total_desc = team_position_salary.sort_values(by='Salary', ascending=False)
team_position_total_desc.columns = ['Team', 'Position', 'Total Salary Spending']
team_position_total_desc.reset_index(drop=True, inplace=True)
team_position_total_desc
```

```python
# Create scatterplot plot for total salary spending across different teams and positions

plt.figure(figsize=(14, 8))
sns.scatterplot(x='Team', y='Total Salary Spending', hue='Position',data=team_position_total_desc)
plt.title('Total Salary Spending Across Teams and Positions')
plt.xlabel('Team')
plt.ylabel('Total Salary Spending')
plt.xticks(rotation=90)
plt.show()
```

```python
# Create bar plot for total salary spending across different teams and positions

plt.figure(figsize=(14, 8))
sns.barplot(x='Team', y='Total Salary Spending', hue='Position', data=team_position_total_desc)
plt.title('Total Salary Spending Across Teams and Positions')
plt.xlabel('Team')
plt.ylabel('Total Salary Spending')
plt.xticks(rotation=90)
plt.show()
```
